Pocket Algo

```python
data = classification.drop(columns=[3])
data = data.rename(columns={4:'label',0:1, 1:2, 2:3})
data[0] = [1] * data.shape[0]

def get_constraints(data, w):
    l = list(range(0,2000))
    random.shuffle(l)
    cons = []
    for row in l:
        s = 0
        for j in range(4):
            s+= w[j] * data[j][row]
    if s < 0:
        if data['label'][row] == 1:
            cons.append(row)
    else:
        if data['label'][row] == -1:
            cons.append(row)
    return cons

def check_constraint(data, w, row):
    s = 0
    for j in range(4):
        s+= w[j] * data[j][row]
    if s < 0:
        if data['label'][row] == 1:
            return 1
    else:
        if data['label'][row] == -1:
            return 2
    return None

import random
def pocket(data, w, alpha, iterations):
    c= []
    cons_min= []
    w_min = w
    for i in range(iterations):
        cons = get_constraints(data,w)
        if len(cons) == 0:
            break
        if check_constraint(data,w,cons[0]) == 1:
            for i in range(4):
                w[i] = w[i] + alpha * data[i][cons[0]]
        else:
            for i in range(4):
                w[i] = w[i] - alpha * data[i][cons[0]]
        if len(cons) < len(cons_min):
            cons_min = cons
            w_min = w
        c.append(len(cons))
#         print(len(c))
    return w_min, c

w = []
for i in range(4):
    w.append(random.uniform(0,1))

result, c = pocket(data, w, 0.5, 7000)
print(result)


def predict(row,w, thr):
    s = 0
    for i in range(4):
        s += row[i] * w[i]
  
    if s >= thr:
        return 1
    else: return -1

data['predicted_label'] = data.apply(lambda x: predict(x,result,-result[0]), axis = 1)
data['compare'] = data.apply(lambda x: (1 if x['label'] == x['predicted_label'] else 0), axis = 1)
accuracy = sum(data['compare'])/data.shape[0]*100
print(accuracy)

# import plotly.express as px
# fig = px.scatter_3d(data, x=1, y=2, z=3,color='label',symbol='label', opacity=0.7)
# fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```

Logistic Regression

```python
import numpy as np
import random
import math
data = classification.drop(columns=[3])
data = data.rename(columns={4:'label',0:1, 1:2, 2:3})
data[0] = [1] * data.shape[0]

def calc(row,w):
    num = []
    for i in range(0,4):
        num.append(row['label'] * row[i])
  
    den = np.matmul(np.matmul([[row['label']]], [w]), [[row[0]], [row[1]],[row[2]], [row[3]]])
    # print(den)
    den = 1 + math.exp(den[0][0])

    return [n/den for n in num]

def gradient(data,w):

    grad = data.apply(lambda row: calc(row,w), axis=1)
    s = [0,0,0,0]
    # print(type(grad), grad[0])
    for i in range(data.shape[0]):
        for j in range(4):
            s[j] += grad[i][j]
  
    for i in range(4):
        s[i] = -s[i]/data.shape[0]
    return s


def logistic_regression(data, w, alpha):

    for i in range(7000):
#         print(i)
        g = gradient(data,w)
        for j in range(4):
            w[j] -= alpha * g[j]

    return w

w = []
for i in range(4):
    w.append(random.uniform(0,1))

result = logistic_regression(data, w, 0.8)
print(result)
```

```python
def predict(row,w):
    s = 0
    for i in range(4):
        s += row[i] * w[i]

    s = math.exp(s)/(1 + math.exp(s))
  
    if s > 0.5:
        return 1
    else: return -1

data['predicted_label'] = data.apply(lambda x: predict(x,result), axis = 1)
data['compare'] = data.apply(lambda x: (1 if x['label'] == x['predicted_label'] else 0), axis = 1)
accuracy = sum(data['compare'])/data.shape[0]*100
print(accuracy)
# import plotly.express as px
# fig = px.scatter_3d(data, x=1, y=2, z=3,color='label',symbol='label', opacity=0.7)
# fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```
